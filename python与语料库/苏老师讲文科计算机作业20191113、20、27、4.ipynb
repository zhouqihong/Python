```python
#1
char = input(":")
letters = []
space = []
number = []
others = []
for i in iter(char):
    if i.isalpha():
        letters.append(i)
    elif i.isspace():
        space.append(i)
    elif i.isdigit():
        number.append(i)
    else:
        others.append(i)
print('''
:{},:{}
:{},:{}
:{},:{}
:{},:{}'''\
      .format(letters,len(letters),space,len(space),number,len(number),others,len(others)))
        
```

```python
#2
char = input(":")
letters = 0
space = 0
number = 0
others = 0
for i in char:
    if i.isalpha():
        letters +=1
    elif i.isspace():
        space +=1
    elif i.isdigit():
        number +=1
    else:
        others +=1
print('=%d'%letters)
print('=%d'%space)
print('=%d'%number)
print('=%d'%others)      
```
