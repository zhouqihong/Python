```python
import pandas as pd
import re
abstract = pd.read_excel(r'D:/NLP//korea_essay.xls',encoding='utf-8')
abstract = abstract.dropna()

data_list = abstract.values.tolist()
for data in data_list:
    title = data[0]
    title = re.sub('[?  / \ < > * | : ]','',title)
    with open(r'D:/NLP///{}.txt'.format(title),'w',encoding='utf-8') as f:
        f.write(data[1])
        f.write(data[2])

```
